#### **1. Money change** 

```python
# Uses python3
import sys

def get_change(n):
    count = 0
    for i in [10, 5, 1]:
        if n>=i:
            q = n//i
            count += q
            n = n%i
        if n==0:
            print(count)
            quit()
    return n

if __name__ == '__main__':
    n = int(sys.stdin.read())
    print(get_change(n))
```

#### **2. Maximum value of the loot** 

```python
# Uses python3
n, W = [int(i) for i in input().split()]
lst = []

if W == 0:
    print(0)
    quit()

for _ in range(n):
    v, w = [int(i) for i in input().split()]
    if v == 0:
        continue
    lst.append((v, w))

lst.sort(key = lambda x: x[0]/x[1], reverse = True)

total_value = 0

for v,w in lst:
    if W==0:
        print(total_value)
        quit()
    amt = min(w, W)
    total_value += amt*v/w
    W -= amt

print(total_value)
```

#### **3. Car fueling** 

```python
import sys

def compute_min_refills(distance, tank, stops):
    stops.append(distance)
    no_stops = 0
    left_dist = distance
    covered_dist = 0
    for i in range(len(stops) - 1):
        if (left_dist == 0):
            break
        elif ((stops[i + 1] - stops[i]) > tank):
            no_stops = -1
            break
        elif ((covered_dist + tank) >= stops[i] and (covered_dist + tank) < stops[i + 1]):
            covered_dist = stops[i]
            left_dist = distance - stops[i]
            no_stops +=  1
    return no_stops


if __name__ == '__main__':
    d, m, _, *stops = map(int, sys.stdin.read().split())
    print(compute_min_refills(d, m, stops))
```

#### **5. Collecting sigantures** 

```python
n = int(input())
lst = []

for _ in range(n):
    a, b = [int(i) for i in input().split()]
    lst.append((a,b))

lst.sort(key = lambda x: x[1])

index = 0
coordinates = []

while index < n:
    curr = lst[index]
    while index < n-1 and curr[1]>=lst[index+1][0]:
        index += 1
    coordinates.append(curr[1])
    index += 1
print(len(coordinates))
print(' '.join([str(i) for i in coordinates]))
```

#### **6. Maximum number of prizes** 

```python
# Uses python3


n = int(input())
if n == 1:
    print(1)
    print(1)
    quit()
W = n
prizes = []
for i in range(1, n):
    if W>2*i:
        prizes.append(i)
        W -= i
    else:
        prizes.append(W)
        break

print(len(prizes))
print(' '.join([str(i) for i in prizes]))
```

#### **7. Maximum salary** 

```python
n = int(input())
lst = [int(i) for i in input().split()]


def IsGreaterOrEqual(digit, max_digit):
    return int(str(digit)+str(max_digit))>=int(str(max_digit)+str(digit))

def largestnumber(lst):
    answer = []
    
    while lst!=[]:
        max_digit = 0
        for digit in lst:
            if IsGreaterOrEqual(digit, max_digit):
                max_digit = digit
        answer.append(max_digit)
        lst.remove(max_digit)

    return answer

print(''.join([str(i) for i in largestnumber(lst)]))
```
