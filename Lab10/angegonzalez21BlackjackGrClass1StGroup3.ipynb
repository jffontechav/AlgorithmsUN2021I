### **Task: Estimate some values**
1. Probality of get the value 21
2. Mean number of cards to get the value 21
---






Let's start to define the different decks to simulate

Here we have the poker deck (52 cards) and the spaish deck (48 cards)

```python
spanish_base = list(range(1,13))
poker_base = list(range(1,14))
```

```python
spanish_deck = 2* spanish_base
poker_deck = 2 * poker_base
print("My spanish deck: {}".format(spanish_deck), len(spanish_deck))
print("My poker deck: {}".format(poker_deck), len(poker_deck))
```

Create our simulation function `simulate_bj`

```python
def simulate_bj(deck_, n, initial_deck = []):
  # as params deck_: type of deck
  #           n: number of random hands   
  count = 0
  number_of_cards = [] 

  for i in range(n):
    deck = list(deck_)
    hand_cards = list(initial_deck)

    while (sum(hand_cards) < 21):
      card = random.choice(deck)
      hand_cards.append(card)
      deck.remove(card) 

    if (sum(hand_cards) == 21):
      count += 1
      number_of_cards.append(len(hand_cards))

  prob = count / n
  mean= sum(number_of_cards) / len(number_of_cards)

  return prob, mean, number_of_cards
```

**With poker we have the following**

```python
# using the poker deck
prob , mean , n_cards = simulate_bj(poker_deck, 10000000)

print('Estimated probability: ', prob)
print('Mean of cards 21: ', mean)
```

```python
#view graphs of probability 

prob_cum= 1-prob
total_prob= [prob, prob_cum]
bins = range(11)

fig, axs = plt.subplots(2, figsize=(9,9))\

# first plot
axs[0].set(title = 'Montecarlo simulation of 21 Black Jack [Poker deck]')
axs[0].hist(n_cards, bins=bins, color= "#6978C9", edgecolor="#111539")
axs[0].set(xlabel = "Number of cards")
axs[0].set(ylabel = "Hands with 21")

#second plot
names = ['success', 'failure']
my_circle = plt.Circle( (0,0), 0.7, color='white')
axs[1].pie(total_prob, labels=names, colors=["#6978C9", "#4A5596"],autopct="%0.2f %%")
axs[1].set(title="Probability (perc.) of get 21")
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

Estimate the profit in a week

```python
def get_profit(d):
  simulations = 10000 # Simulations 
  games = 600 # Games in a week
  
  count = 0 
  number_of_cards = []

  bet = 0 
  acum_profit = 0
  profit_list = list()
  prob_21 = list()

  possible_games = [1] + 9*[0] 

  for i in range(simulations):
    count = 0
    for game in range(games):
      choice = random.choice(possible_games)
      acum_profit = 0   
      if choice:
        hand_cards = [7, 7, 5] 
        deck = 4*([1]+[2])
        bet = 1000
      else:
        hand_cards = []
        deck = list(d)
        bet = 10

      while (sum(hand_cards) < 21):
        card = r.choice(deck)
        hand_cards.append(card)
        deck.remove(card)

      if sum(hand_cards) == 21:
        count += 1
        number_of_cards.append(len(hand_cards))
        acum_profit += bet
      else:
        acum_profit -= bet

    profit_list.append(acum_profit)
    prob_of_21 = count / games
    prob_21.append(prob_of_21)
    mean_of_21 =(sum(prob_21) / len(prob_21))
    mean_cards = sum(number_of_cards) / len(number_of_cards)
  
  return profit_list,mean_of_21, mean_cards, number_of_cards
```
