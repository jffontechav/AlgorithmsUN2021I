#### **1. Fibonacci Number**

```python
from math import sqrt

# Uses python3


def calc_fib(n):
    if (n <= 1):
        return n

    return calc_fib(n - 1) + calc_fib(n - 2)


def fib_list(n):

    if n == 0:
        return 0
    if n == 1:
        return 1

    fib = [x for x in range(n+1)]
    fib[0] = 0
    fib[1] = 1

    for i in range(2, n+1):
        fib[i] = fib[i-1]+fib[i-2]

    return fib[n]


def fib_ultra(n):
    aureo = (1+sqrt(5))/2
    return int((aureo ** n - (1-aureo)**n)/sqrt(5))
```

####  **3. Greatest common divisor (GCD)**

```python
# Uses python3

def gcd_naive(a, b):
    current_gcd = 1
    for d in range(2, min(a, b) + 1):
        if a % d == 0 and b % d == 0:
            if d > current_gcd:
                current_gcd = d

    return current_gcd


def get_gcd_ec(a, b):
    if b == 0:
        return a
    a_prime = a % b
    return get_gcd_ec(b, a_prime)

a, b = map(int, input().split())
print(get_gcd_ec(a, b))
```

#### **4. Least common multiple (LCM)**

```python
# Uses python3
import sys


def lcm_naive(a, b):
    for l in range(1, a*b + 1):
        if l % a == 0 and l % b == 0:
            return l

    return a*b


def get_gcd_ec(a, b):
    if b == 0:
        return a
    a_prime = a % b
    return get_gcd_ec(b, a_prime)


def get_lcm(a, b):
    return int((a/get_gcd_ec(a, b)) * b)


a, b = map(int, input().split())
print(get_lcm(a, b))
```
